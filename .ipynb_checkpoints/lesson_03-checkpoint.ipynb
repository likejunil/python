```python
import FinanceDataReader as fdr
import pandas as pd
```

```python
# 하나의 cell 에서 multiple output 출력을 가능하게 하는 코드
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df1 = fdr.DataReader("005930", "2018-01-01", "2018-12-31")
df1.shape
df1
```

```python
# DataFrame 에서 특정 index 를 삭제할 수 있다.
# 데이터가 없더라도 index 자체가 생략되면 안 된다.
# 이런 경우 정상 DataFrame 의 index 를 가져와서..
# reindex 를 해주면 없어진 index 를 살릴 수 있다.
# index 를 살리면 nan 가 채워진다.
# nan 를 전략에 따라 임의의 값으로 채워야 한다.

df2 = df1.drop(pd.to_datetime("2018-01-04"))
df2.head()
df2 = df2.reindex(df1.index)
df2.head()
df2.fillna(method='ffill').head()
```
