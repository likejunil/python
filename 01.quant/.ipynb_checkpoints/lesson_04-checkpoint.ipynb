### describe()

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
df.describe()
```

```python
# 수치값들을 대상으로 개수, 평균, 편차, 최소, 최대, 4분위 등을 보여준다.
# 수치값이 아닌 Series 들을 drop 한다.
# include 옵션을 통해 어떤 Series 들을 대상으로 할 것인지 지정할 수 있다.

# df.describe().T 
df.describe(include=[np.int64, np.float64]).T.head()
```

```python
# 4분위가 아닌 임의의 퍼센트를 지정하여 볼 수 있다.

df.describe(percentiles=[0.01, 0.1, 0.9]).T.head()
```

```python
# describe() 의 대상을 수치가 아닌 값으로 할 경우..
# 가장 빈도수가 높은 항목에 대한 정보를 확인할 수 있다.
# 혹은 특정 데이터의 타입을 제외할 수도 있다.

df.describe(include=[pd.Categorical]).T
df.describe(exclude=[np.number]).T
```

### unique(), value_counts()

```python
# Series.unique() 는 중복되지 않게 원소들을 하나씩 선택하여 numpy.ndarray 를 반환한다.
# Series.unique() 는 nan 를 포함한다.
# DataFrame.nunique() 는 각 Series 에 대하여 유일한 값의 데이터들을 센다.
# DataFrame.nunique() 는 nan 를 무시하고 센다.

a1 = pd.Series([1, 1, 3], index=['a', 'b', 'c'])
a2 = pd.Series([4, 5, 6], index=['a', 'b', 'd'])
df1 = pd.DataFrame({'r1':a1, 'r2':a2})
df1
df1.nunique()
df.nunique()
```

```python
# Series 에는 unique(), nunique(), value_counts() 3가지를 사용할 수 있다.

a1 = pd.Series([1, 1, 3, np.nan], index=['a', 'b', 'c', 'd'])
a2 = pd.Series([4, 5, 6, 7], index=['a', 'b', 'd', 'e'])
df1 = pd.DataFrame({'r1':a1, 'r2':a2})

a1.unique() # nan 를 센다.
a1.nunique() 
a1.value_counts()

df1.nunique()
df1.value_counts()
```
